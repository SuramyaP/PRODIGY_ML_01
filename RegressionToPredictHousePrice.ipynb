```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df.columns
```

```python
nf = df[['LotArea', 'BedroomAbvGr','BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'SalePrice']].copy()
```

```python
nf
```

```python
attributes_to_sum = ['BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath']

# Create a new column 'TotalBathrooms' with the sum of the selected attributes
nf['TotalBathrooms'] = df[attributes_to_sum].sum(axis=1)
```

```python
nf
```

```python
columns_to_drop = ['BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath']

nf.drop(columns=columns_to_drop, inplace=True)
```

```python
nf
```

```python
df_test = pd.read_csv('test.csv')
```

```python
df_test
```

```python
nf_test = df_test[['LotArea', 'BedroomAbvGr','BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath']].copy()
```

```python
nf_test['TotalBathrooms'] = nf_test[attributes_to_sum].sum(axis=1)
```

```python
nf_test
```

```python
nf_test.drop(columns=columns_to_drop, inplace=True)
```

```python
nf_test
```

```python
nf_data = nf[['LotArea', 'BedroomAbvGr','TotalBathrooms']].copy()
```

```python
nf_target = nf.drop(columns=['LotArea', 'BedroomAbvGr', 'TotalBathrooms'])
```

```python
model = LinearRegression()
```

```python
model.fit(nf_data, nf_target)
```

```python
predictions = model.predict(nf_test)
```

```python
nf_test_target = pd.read_csv('sample_submission.csv')
```

```python
nf_test_target
```

```python
nf_test_tar = nf_test_target['SalePrice'].values
```

```python
nf_test_tar
```

```python
mse = mean_squared_error(nf_test_tar, predictions )
```

```python
mse
```

```python
r2 = r2_score(nf_test_tar, predictions )
```

```python
r2
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(nf_test_tar, predictions)
mae
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
rmse = np.sqrt(mean_squared_error(nf_test_tar, predictions))
rmse
```
